```python
# Step 1 - ChatGPT prompt:
# Write code for a google colab notebook in python.
# Make a matrix 250 pixels by 250 pixels in grayscale, this will be the image background, make it black.
# Make a square 50 x 50 pixels in white, and place it in the center of the background square.
# Plot in the screen these result.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a black background matrix
background = np.zeros((250, 250), dtype=np.uint8)

# Calculate the position to place the white square in the center
center_x = background.shape[1] // 2
center_y = background.shape[0] // 2

# Create a white square of size 50x50
white_square = np.ones((50, 50), dtype=np.uint8) * 255

# Place the white square in the center of the background
background[center_y - 25 : center_y + 25, center_x - 25 : center_x + 25] = white_square

# Plot the result
plt.imshow(background, cmap='gray', vmin=0, vmax=255)
plt.axis('off')  # Turn off axis labels
plt.show()
```
