```python
# Step 3c - Plot sample images from each class.
#         a simple utility to visualize 25 images from a directory

# ==============================================================================
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# -----------------------------------------------------------------------------
# Replace these folder paths with your actual folder paths
folder_paths = ["/content/birds",
                "/content/drones",
                "/content/others"]
# -----------------------------------------------------------------------------
def select_random_images(folder_path, num_images=25):
    all_images = os.listdir(folder_path)
    selected_images = random.sample(all_images, num_images)
    return selected_images
# -----------------------------------------------------------------------------
def display_images(images, folder_path):
    fig, axes = plt.subplots(5, 5, figsize=(10, 10))
    fig.suptitle(f"Randomly Selected Images from {folder_path}", fontsize=16)
    for i in range(5):
        for j in range(5):
            image_name = images[i * 5 + j]
            image_path = os.path.join(folder_path, image_name)
            img = mpimg.imread(image_path)

            axes[i, j].imshow(img)
            axes[i, j].axis('off')
            axes[i, j].set_title(f"{image_name}\n{img.shape[1]} x {img.shape[0]}")

    plt.show()
# -----------------------------------------------------------------------------
for folder_path in folder_paths:
    selected_images = select_random_images(folder_path)
    display_images(selected_images, folder_path)
    print(f"Selected Images from {folder_path}:\n{selected_images}\n")
# ==============================================================================
```

```python
# Step 4a - Make a dataframe and a csv file with the dataset
#           files paths and classes
#           (for later use with Tensorflow)

# modifying and generalizing code from exercise_day#4:

# ==============================================================================
import os
import pandas as pd
# ------------------------------------------------------------------------------
#                                       define the name of the dataset csv file.
dataset_csv = 'dataset_drones.csv'
# ------------------------------------------------------------------------------
#                                     make a list with the names of the columns.
col_lst=['Path', 'Label']
#
# ------------------------------------------------------------------------------
#           make a list with the paths and their corresponding labels (classes).
paths = ['/content/drones/',
         '/content/birds/',
         '/content/others/']
labels = [0, 1, 2]
# ------------------------------------------------------------------------------
# ------------------------------------------------------------------------------
#                                                          define an empty list.
df1_lst = []
# iterate through the folders and files to populate the DataFrame
for label, path in zip(labels, paths):
    file_list = os.listdir(path)
    for filename in file_list:
        # Construct the full file path
        file_path = os.path.join(path, filename)
        # append to the list
        element = (file_path, label)
        df1_lst.append(element)
    # --------------------------------------------------------------------------
# ------------------------------------------------------------------------------
#                                make a dataframe with the list we just created.
df_dataset = pd.DataFrame(df1_lst, columns=col_lst)
# ------------------------------------------------------------------------------
#                                           write the dataframe into a csv file.
df_dataset.to_csv(dataset_csv, index = False)
# ==============================================================================
```

```python
# Step 4a - Convert processed images into a TensorFlow dataset.
#      4b - Split the dataset into train set and test set.

# ==============================================================================
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split
# ------------------------------------------------------------------------------
#                                         read the dataset descriptive csv file.
df_dataset = pd.read_csv('/content/dataset_drones.csv')
# ------------------------------------------------------------------------------
#                                                load and preprocess the images.
def load_and_preprocess_image(image_path, label):
    img = load_img(image_path, target_size=(32, 32))
    img_array = img_to_array(img) / 255.0  # Normalize pixel values to [0, 1]
    return img_array, label
# ------------------------------------------------------------------------------
#                               apply the function to each row in the DataFrame.
image_data = [
    load_and_preprocess_image(row['Path'], row['Label'])
    for _, row in df_dataset.iterrows()
]
# ------------------------------------------------------------------------------
#                                      separate the data into images and labels.
images, labels = zip(*image_data)
# ------------------------------------------------------------------------------
#                                             convert the lists to NumPy arrays.
images = np.array(images)
labels = np.array(labels)
# ------------------------------------------------------------------------------
#                                 split the data into training and testing sets.
train_images, test_images, train_labels, test_labels = train_test_split(
    images, labels, test_size=0.2, random_state=42
)
# ------------------------------------------------------------------------------
#                                                    create TensorFlow datasets.
train_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_labels))
test_dataset = tf.data.Dataset.from_tensor_slices((test_images, test_labels))
# ------------------------------------------------------------------------------
#                                                shuffle and batch the datasets.
batch_size = 32
train_dataset = train_dataset.shuffle(len(train_images)).batch(batch_size)
test_dataset = test_dataset.batch(batch_size)
# ==============================================================================
```

** Step 4c - Use Keras to build a simple CNN classifier**

(How many perceptrons should the output layer have?)

prompt for ChatGPT:

Now, use keras to build a simple CNN classifier.
We will have 3 outputs: drone, bird, other

```python
# Step 4c - Using keras, make a simple CNN classifier.

# ==============================================================================
import tensorflow as tf
from tensorflow.keras import layers, models
# ------------------------------------------------------------------------------
# Define the model
model = models.Sequential()

# Add a Convolutional layer with 32 filters, a 3x3 kernel, and 'relu' activation
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
# Add a MaxPooling layer with 2x2 pool size
model.add(layers.MaxPooling2D((2, 2)))

# Add another Convolutional layer with 64 filters and a 3x3 kernel
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
# Add another MaxPooling layer
model.add(layers.MaxPooling2D((2, 2)))

# Add a third Convolutional layer with 64 filters and a 3x3 kernel
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Flatten the output before the dense layers
model.add(layers.Flatten())

# Add a Dense layer with 64 units and 'relu' activation
model.add(layers.Dense(64, activation='relu'))

# Add the output layer with 3 units (for 'drone', 'bird', 'other') and 'softmax' activation
model.add(layers.Dense(3, activation='softmax'))
# ------------------------------------------------------------------------------
#                                                             compile the model.
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
# ------------------------------------------------------------------------------
#                                                     display the model summary.
model.summary()
# ==============================================================================
```

** Step 4d - Train the model**

 and evaluate performance; Display the accuracy of the trained model on the test set.


```python
# Step 4d - Train the model

# ==============================================================================
#                                                   define the number of epochs.
epochs = 20
# ------------------------------------------------------------------------------
#                                   train the model using the prepared datasets.
history = model.fit(train_dataset, epochs=epochs, validation_data=test_dataset)
# ==============================================================================
```
